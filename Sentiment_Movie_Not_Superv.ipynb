##Actividad 4 Modelo No Supervisado

Objetivo: Buscamos determinar qué modelo de aprendizaje automático es el más apropiado para predecir el sentimiento (output) basado en una crítica de película (input).

Input(x) -> movie review
Output(y) -> sentiment

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans


```

Leer Dataset

```python
df_review = pd.read_csv('revisiones_peliculas.csv')

df_review
```

```python
# Supongamos que tienes un dataframe df_review con una columna 'review' que contiene las revisiones

# Vectorización de texto
tfidf = TfidfVectorizer(stop_words='english')
review_vectors = tfidf.fit_transform(df_review['review'])

# Construcción del modelo de K-means
k = 2  # Número de clusters deseado
kmeans = KMeans(n_clusters=k)
kmeans.fit(review_vectors)

# Asignación de cada instancia a un cluster
clusters = kmeans.labels_

```

```python
# Agregar los clusters al dataframe
df_review['cluster'] = clusters

# Mostrar los resultados
print("Distribución de instancias en cada cluster:")
print(df_review['cluster'].value_counts())

# Puedes explorar los clusters y analizar las revisiones dentro de cada uno para encontrar patrones o temas comunes
for cluster_id in range(k):
    print(f"\nCluster {cluster_id}:")
    reviews_in_cluster = df_review[df_review['cluster'] == cluster_id]['review']
    for review in reviews_in_cluster.sample(min(5, len(reviews_in_cluster))):  # Mostrar hasta 5 revisiones
        print(review)
```
